```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
from sklearn import svm
from sklearn.linear_model import SGDClassifier
```

### Read in csvs and store into dataframes according to year

```python
dfs = dict()
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
for year in years:
    dfs[year] = pd.read_csv(f'data/{year}_school_data.csv')
```

### Create new dataframe of all years except 2017

```python
df_list = [dfs[year] for year in years[:-1]]
df_agr = pd.concat(df_list)
```

### Split new dataframe values to x and y

```python
x = df_agr[['total_students', 'reduced_price']].values
y = df_agr.title_I.values 
print(x.shape, y.shape)
```

### Split yearly dataframe values to x and y for testing

```python
x_vals = dict()
y_vals = dict()
for year, df in dfs.items():
    x_vals[year] = df[['total_students', 'reduced_price']].values
    y_vals[year] = df.title_I.values 
    print(y_vals[year].shape, x_vals[year].shape)
```

# Linear SVC
    Data is already in standard scalar format, handled in EDA

```python
lin_clf = svm.LinearSVC(C = 1e-5) 
lin_clf.fit(x, y)
```

### Test on 2017 data gathered

```python
lin_score = lin_clf.score(x_vals[2017],y_vals[2017])
print(f'Linear SVC score: {lin_score}')
```

### Confusion matrix plot

```python
cm = confusion_matrix(y_vals[2017], lin_clf.predict(x_vals[2017]))
plt.figure()
plot_confusion_matrix(cm, figsize=(12,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Decision Tree Model - Confusion Matrix")
plt.xticks(range(2), ["Class 1","Class 2"], fontsize=16)
plt.yticks(range(2), ["Class 1","Class 2"], fontsize=16)
plt.show()
```

Linear SVC plot function

```python
def plot_lin_svc(X, y, clf, C = 1e-5):
    plt.figure(figsize=(30, 15))
    decision_function = clf.decision_function(X)
    support_vector_indices = np.where(
        np.abs(decision_function) <= 1 + 1e-15)[0]
    support_vectors = X[support_vector_indices]

    plt.subplot(1, 2, 1)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                         np.linspace(ylim[0], ylim[1], 50))
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
                linestyles=['--', '-', '--'])
    plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=100,
                linewidth=1, facecolors='none', edgecolors='k')
    plt.title(f'C={str(C)} support vectors')
    plt.tight_layout()
    plt.show()
```

```python
plot_lin_svc(x, y, lin_clf)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

# Decision Tree

```python
#create decision tree regressor
clf_1 = DecisionTreeClassifier(max_leaf_nodes=5, random_state=30, max_depth=1,criterion='entropy')
clf_2 = DecisionTreeClassifier(max_leaf_nodes=10, random_state=30, max_depth=2,criterion='entropy')
clf_3 = DecisionTreeClassifier(max_leaf_nodes=15, random_state=30, max_depth=5,criterion='entropy')
clf_4 = DecisionTreeClassifier(max_leaf_nodes=25, random_state=30, max_depth=25,criterion='entropy')

#fit regression model
clf_1.fit(x, y)
clf_2.fit(x, y)
clf_3.fit(x, y)
clf_4.fit(x, y)

#predict
y_1 = clf_1.predict(x_vals[2017])
y_2 = clf_2.predict(x_vals[2017])
y_3 = clf_3.predict(x_vals[2017])
y_4 = clf_4.predict(x_vals[2017])

#calculate mean_squared_error
mse_1 = mean_squared_error(y_vals[2017], y_1)
mse_2 = mean_squared_error(y_vals[2017], y_2)
mse_3 = mean_squared_error(y_vals[2017], y_3)
mse_4 = mean_squared_error(y_vals[2017], y_4)

#print mse
print(f'Mean Squared Error 1: {mse_1}')
print(f'Mean Squared Error 2: {mse_2}')
print(f'Mean Squared Error 3: {mse_3}')
print(f'Mean Squared Error 4: {mse_4}')

#find the score
clf_1_score = clf_1.score(x_vals[2017],y_vals[2017])
clf_2_score = clf_2.score(x_vals[2017],y_vals[2017])
clf_3_score = clf_3.score(x_vals[2017],y_vals[2017])
clf_4_score = clf_4.score(x_vals[2017],y_vals[2017])

#print score
print(f'Decision Tree 1 score: {clf_1_score}')
print(f'Decision Tree 2 score: {clf_2_score}')
print(f'Decision Tree 3 score: {clf_3_score}')
print(f'Decision Tree 4 score: {clf_4_score}')
```

### Confusion Matrix

```python
cm = confusion_matrix(y_vals[2017], y_1)
plt.figure()
plot_confusion_matrix(cm, figsize=(12,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Decision Tree Model - Confusion Matrix")
plt.xticks(range(2), ["Class 1","Class 2"], fontsize=16)
plt.yticks(range(2), ["Class 1","Class 2"], fontsize=16)
plt.show()
```

### K-fold cross validation

```python
from sklearn.model_selection import KFold, cross_val_score

scores = cross_val_score(clf_4,x_vals[2017],y_vals[2017],cv=5)
print("Cross-validation scores: {}".format(scores))
print("Average cross-validation score: {:.2f}".format(scores.mean()))
```

### Decision Tree with Bagging

```python
from sklearn.ensemble import BaggingClassifier

bagged_clf = BaggingClassifier(clf_1, n_estimators=20, random_state=1)
bagged_clf.fit(x,y)
pred = bagged_clf.predict(x_vals[2017])
bagged_mse = mean_squared_error(y_vals[2017], pred)
print(f'Bagged Mean Squared Error: {bagged_mse}')

bagged_clf_score = bagged_clf.score(x_vals[2017],y_vals[2017])
print(f'Bagged Decision Tree score: {bagged_clf_score}')
```
